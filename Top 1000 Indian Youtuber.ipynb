```python
df_new=df.groupby(['category'])['video views(in millions)'].sum()
df_new=df_new.sort_values()
df_new
def my_color_shade():
    colors=[]
    for i,j,k in zip(range(75,100,10),range(0,100,15),range(100,0,-15)):
        colors.append((i/100,j/100,k/100,0.5))
    return colors
        
    
plt.barh(df_new.keys(),df_new,color=my_color_shade())
for i,j in enumerate(df_new):
    
    plt.text(j,i-0.35,str(j))
# plt.show()
```
